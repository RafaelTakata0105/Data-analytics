# Intro to predictive models

```python
from sklearn.datasets import make_regression
from sklearn.linear_model import LinearRegression
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
x, y = make_regression(n_features= 1, noise= 30, random_state= 0)
df = pd.DataFrame(x)
df.columns = ['x']
df['y'] = y
df
```

```python
df['x'].plot(kind='hist')
```

```python
y_mean = np.array([y.mean()]*100)
y_mean[0] #No importa porque todos son el mismo
```

```python
model_1 = LinearRegression()
model_2 = LinearRegression()
model_3 = LinearRegression()

model_1.fit(x, y)
```

```python
y_pred = model_1.predict(x)
```

```python
#Selecetor
plt.figure(figsize=(10,3))
#Datos a graficar
plt.scatter(x, y, label = 'Random Data')
#Graficar el promedio
plt.plot(x, y_mean, label = 'Mean', c = 'red')
#Graficar el linear regression
plt.plot(x, y_pred, label = 'linear regression', c = 'black')
#Formato del show
plt.title('Scatter Example plot')
plt.xlabel('Feature')
plt.ylabel('Target')
plt.legend()

plt.savefig('../reports/figures/random.svg')
```
